# HEPATITIS C PREDICTION APPLICATION
Hepatitis adalah infeksi yang menyerang organ hati, dan ada tiga jenis hepatitis, yaitu A, B, dan C. Pengobatan untuk hepatitis C sulit dilakukan karena ketersediaan obat yang terbatas, dan pasien harus menghadapi biaya yang tinggi. Oleh karena itu, sangat penting untuk segera menangani penyakit ini dengan serius.

```python
pd.options.display.max_rows = 9999

df = pd.read_csv("dataset/HepatitisCdata.csv")
df_mod = pd.read_csv("dataset/HepatitisCdata.csv")
```

```python
ALL = 'ALL'
def unique_sorted_values_plus_ALL(array):
    unique = array.unique().tolist()
    unique.sort()
    unique.insert(0, ALL)
    return unique
```

```python
# read data

output_read_data = widgets.Output()

dropdown_age = widgets.Dropdown(
    options = unique_sorted_values_plus_ALL(df.Age), 
    description="Age :"
)
dropdown_category = widgets.Dropdown(
    options = unique_sorted_values_plus_ALL(df.Category), 
    description="Category :"
)
max_row_picker = widgets.IntSlider(
    value=1,
    min=0,
    max=len(df["Category"]),
    step=1,
    description='Display Row :',
    disabled=False,
    continuous_update=False,
    orientation='horizontal'
)

def filtering(age, category, max_row):
    output_read_data.clear_output()
    with output_read_data:
        if (age == ALL) & (category == ALL):
            showdf = df.head(max_row)
            filtering = showdf

        elif (age == ALL) & (max_row > 0):
            showdf = df[(df.Category == category)]
            filtering = showdf.head(max_row)
            
        elif (category == ALL) & (age != ALL) & (max_row > 0):
            showdf = df[(df.Age == age)]
            filtering = showdf.head(max_row)

        elif (category == ALL) & (max_row > 0):
            showdf = df[(df.Age == age)]
            filtering = showdf.head(max_row)
            
        elif (age == ALL) & (category != ALL) & (max_row > 0):
            showdf = df[(df.Age == age)]
            filtering = showdf.head(max_row)
        
        elif (age != ALL) & (category != ALL) & (max_row > 0):
            showdf = df[(df.Age == age) & (df.Category == category)]
            filtering = showdf.head(max_row)
            
#         else:
#             showdf = df[(df.Age == age) & (df.Category == category)]
#             filtering = showdf.head(max_row)
    
        display(filtering)
def dropdown_age_handler(change):
    filtering(change.new, dropdown_category.value, max_row_picker.value)

def dropdown_category_handler(change):
    filtering(dropdown_age.value, change.new, max_row_picker.value)
    
def max_row_picker_handler(change):
    filtering(dropdown_age.value, dropdown_category.value, change.new)
    
dropdown_age.observe(dropdown_age_handler, names='value')
dropdown_category.observe(dropdown_category_handler, names='value')
max_row_picker.observe(max_row_picker_handler, names='value')
```
